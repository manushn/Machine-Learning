```python
import numpy as np 
import pandas as pd 
from sklearn.linear_model import LinearRegression
```

```python
train_data=pd.read_csv("data.csv")
```

```python
train_data.head()
```

```python
train_data.isnull().sum()
```

```python
train_data.info()
```

```python
train_data["city"].unique()
```

```python
train_final_data=train_data.drop(columns=['date', 'street', 'country'])
train_final_data=pd.get_dummies(train_final_data,columns=['city', 'statezip'], drop_first=True)
```

```python
train_final_data.info()
```

```python
X_train=train_final_data.drop('price',axis=1)
y_train=train_final_data['price']
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
test_data=pd.read_csv("output.csv")
```

```python
test_data.info()
```

```python
test_final_data=test_data.drop(columns=['date', 'street', 'country'])
test_final_data=pd.get_dummies(test_final_data,columns=['city', 'statezip'], drop_first=True)
```

```python
X_test=test_final_data.drop('price',axis=1)
y_test=test_final_data['price']
```

```python
from sklearn.metrics import r2_score
```

```python
y_predict=model.predict(X_test)
```

```python
score=r2_score(y_test,y_predict)
```

```python
print(f"The Accuracy on un seen data {score}")
```

```python
y_seen_predict=model.predict(X_test)
```

```python
seenscore=r2_score(y_train,y_seen_predict)
```

```python
print(f"The Accuracy on seen data {seenscore}")
```
